Pie chart showing the market share of mobile OS

```python
import matplotlib.pyplot as plt

labels = ['Android', 'iOS', 'Others']
market_share = [72, 26, 2]
colors = ['green', 'red', 'orange']

plt.pie(market_share, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title("Market Share of Mobile Operating Systems")
plt.legend(title="Mobile OS", loc='upper left')
plt.axis('equal')  # Makes the pie a circle
plt.tight_layout()
plt.show()
```

Horizontal bar chart showing the market share of mobile OS

```python
import matplotlib.pyplot as plt

labels = ['Android', 'iOS', 'Others']
market_share = [72, 26, 2]
colors = ['green', 'red', 'orange']

plt.barh(labels, market_share, color=colors)
plt.ylabel("Mobile Operating Systems")
plt.xlabel("Market Share (%)")
plt.title("Market Share of mobile Operating System (Bar Chart)")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
