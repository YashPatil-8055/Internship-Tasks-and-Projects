Side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag

```python
import matplotlib.pyplot as plt
import numpy as np

years = ['2004', '2006', '2008', '2010', '2012']

sachin_runs = [1100, 900, 1200, 850, 950]
kohli_runs = [0, 300, 1000, 1250, 1050]
sehwag_runs = [950, 800, 1100, 980, 700]

# Setting position of bars on X axis
x = np.arange(len(years))
bar_width = 0.25

plt.bar(x - bar_width, sachin_runs, width=bar_width, label='Sachin', color='blue')
plt.bar(x, kohli_runs, width=bar_width, label='Kohli', color='green')
plt.bar(x + bar_width, sehwag_runs, width=bar_width, label='Sehwag', color='red')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs of Sachin, Kohli & Sehwag Over 5 Years')
plt.xticks(x, years)
plt.legend()
plt.show()
```

Horizontal bar chart showing total runs scored in their debut 5 years

```python
players = ['Sachin', 'Kohli', 'Sehwag']
debut_runs = [5200, 4800, 4400]

plt.barh(players, debut_runs, color=['orange', 'black', 'green'])

plt.ylabel("Players")
plt.xlabel("Total Runs")
plt.title("Runs in Debut 5 Years")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
