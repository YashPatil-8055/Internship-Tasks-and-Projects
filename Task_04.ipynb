Displaying 10 Different Matplotlib Styles

```python
import matplotlib.pyplot as plt

students = ['Yash', 'Jagdish', 'Paresh', 'Shreeraj']
marks = [45, 40, 42, 40]
plt.bar(students, marks, color=['orange', 'blue', 'green', 'red'])
plt.title("Bar Plot: Student's Marks in Python Test")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```

```python
matches = [1, 2, 3, 4, 5]
scores = [45, 67, 89, 70, 92]

plt.plot(matches, scores, linestyle='--', marker='o', color='green')
plt.title("Line Plot: Cricketer's Scores Over 5 Matches")
plt.xlabel("Match Number")
plt.ylabel("Runs Scored")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
height = [150, 160, 165, 170, 180]
weight = [50, 60, 65, 68, 75]

plt.scatter(height, weight, color='purple')
plt.title("Scatter Plot: Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
activities = ['Study', 'Sleep', 'Play', 'Screen Time', 'Others']
hours = [6, 8, 3, 5, 2]

plt.pie(hours, labels=activities, autopct='%1.1f%%')
plt.title("Pie Chart: Daily Time Distribution")
plt.tight_layout()
plt.show()
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
study = [2, 3, 4, 3, 5]
play = [1, 1, 2, 2, 1]
screen = [3, 2, 2, 3, 4]

plt.stackplot(days, study, play, screen, labels=['Study', 'Play', 'Screen'], colors=['blue', 'green', 'red'])
plt.title("Stack Plot: Time Spent on per Activity (Weekdays)")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.legend(loc='upper left')
plt.show()
```

```python
subjects = ['Math', 'Science', 'English', 'History']

yash = [85, 90, 78, 88]
jagdish = [80, 88, 83, 85]

plt.plot(subjects, yash, marker='o', label='Yash')
plt.plot(subjects, jagdish, marker='s', label='Jagdish')
plt.title("Line Plot: Subject-wise Marks Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
teams = ['MI', 'CSK', 'KKR', 'RR', 'SRH',  'RCB']
titles = [5, 5, 2, 1, 1, 1]

plt.barh(teams, titles, color='gold')
plt.title("Horizontal Bar Chart: IPL Titles by Team")
plt.xlabel("Number of Titles")
plt.ylabel("Teams")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('dark_background')

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.plot(x, y, marker='o', linestyle='-', color='cyan')
plt.title("Line Plot with Dark Background Style")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.tight_layout()
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
product_A = [40, 35, 30, 35, 27]
product_B = [25, 32, 34, 20, 25]

plt.fill_between(months, product_A, color="skyblue", alpha=0.5, label='Product A')
plt.fill_between(months, product_B, color="salmon", alpha=0.5, label='Product B')
plt.title("Area Plot: Monthly Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Sales (in 1000s)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

semesters = ['Sem 1', 'Sem 2', 'Sem 3', 'Sem 4']
yash_marks = [71, 80, 72, 86]

plt.step(semesters, yash_marks, linewidth=2, color='green', marker='o')
plt.title("Yash's Semester Performance")
plt.xlabel("Semester")
plt.ylabel("Percentage")
plt.tight_layout()
plt.show()
```
